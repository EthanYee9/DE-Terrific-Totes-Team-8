```python
dim_date = pd.read_parquet('../tempdata/parquets_for_viz/dim_date_15-09-05.295403.parquet')
dim_location = pd.read_parquet('../tempdata/parquets_for_viz/dim_location_15-09-05.295403.parquet')
dim_design = pd.read_parquet('../tempdata/parquets_for_viz/dim_design_15-09-05.295403.parquet')
dim_staff = pd.read_parquet('../tempdata/parquets_for_viz/dim_staff_15-09-05.295403.parquet')
dim_currency = pd.read_parquet('../tempdata/parquets_for_viz/dim_currency_15-09-05.295403.parquet')
dim_counterparty = pd.read_parquet('../tempdata/parquets_for_viz/dim_counterparty_15-09-05.295403.parquet')
fact_sales_order = pd.read_parquet('../tempdata/parquets_for_viz/fact_sales_order_15-09-05.295403.parquet')
```

```python
#location and sales
location_sales = pd.merge(fact_sales_order, dim_location, left_on = 'agreed_delivery_location_id', right_on = 'location_id')
sns.boxplot(data=location_sales, x="city", y="units_sold")
plt.xticks(rotation=90)
```

```python
# date and sales

dim_date['date_id'] = dim_date['date_id'].astype(str)
dates_sales = pd.merge(fact_sales_order, dim_date, left_on = 'agreed_delivery_date', right_on = 'date_id')
dates_sales['total_sale'] = dates_sales['unit_price'] * dates_sales['units_sold']
sns.boxplot(data=dates_sales, x="day_name", y="total_sale", hue ="quarter")
plt.xticks(rotation=90)
```

```python
#sales and staff

staff_sales = pd.merge(dates_sales, dim_staff, left_on = 'sales_staff_id', right_on = 'staff_id')
staff_sales['total_sale'] = staff_sales['unit_price'] * staff_sales['units_sold']
sns.barplot(data=staff_sales, x="first_name", y="total_sale", hue="quarter")
plt.xticks(rotation=90)
```
